BARPLOT
  </h2>

A barplot is a chart that represents data with rectangular bars where the height of each bar is proportional to the value it represents. It’s used to compare values between different categories or groups. It provides a clear visual representation of the frequency, mean, or other statistics for each category.

<b>CATEGORICAL DATA</b>

<b>NUMERICAL DATA</b>

```python
sns.barplot(x='day', y='total_bill', hue='sex', data=tips, palette = 'pastel',
           edgecolor=".2", errcolor=".2")
plt.xticks(fontsize=14) # Размер подписей на оси X
plt.yticks(fontsize=14)
plt.title('Average score by day of the week, separated by gender', fontsize=16)
sns.despine() 
plt.show()
```

BOXPLOT
  </h2>
</div>

A boxplot (also known as a box and whisker plot) is a standardized way of graphically depicting the distribution of numerical data. It displays the median, quartiles (25th and 75th percentiles), and outliers. A boxplot helps understand the spread of the data, its symmetry, and the presence of any anomalies.

<b>NUMERICAL DATA</b>

```python
sns.boxplot(x='time', y='tip', data=tips,
           palette="light:b", linewidth = 1.2,  fliersize=5, 
            whis=1.5)
sns.despine()
plt.tight_layout()
plt.show()
```

RELPLOT
  </h2>
</div>

Relplot (short for relational plot) in Seaborn is a high-level function for creating plots that show the relationships between two or more numerical variables. It allows you to easily create various plot types, such as scatterplot and lineplot, with additional capabilities for grouping and visualizing additional categorical variables. relplot simplifies plot creation by hiding some of the implementation details that would be required when using lower-level matplotlib functions.

<b>X, Y - NUMERICAL DATA</b>

<b>hue (optional) - CATEGORICAL DATA
</b>

```python
sns.relplot(x='total_bill', y='tip', col='sex', hue = 'time',
            sizes=(20, 150), data=tips, palette='magma'
            )
plt.title('Tips are based on the size of the bill and gender')
plt.show()
```

SCATTERPLOT
  </h2>
</div>

A scatterplot is a type of plot that displays the relationship between two numerical variables. Each point on the plot represents a single observation, with the point’s position on the plane determined by the values of these two variables. A scatterplot helps to visually assess whether there is any correlation or pattern between the variables and to see the presence of clusters, outliers, or other features in the data distribution.

<b>TWO NUMERICAL DATA</b>

```python
markers = {'Male': 'o', 'Female': 's'}
scatter = sns.scatterplot(x='total_bill', y='tip', hue='sex', s=100, alpha=0.8, data=tips)
plt.grid(True, alpha=0.5)
plt.title('The dependence of tips on the amount of the bill, taking into account gender', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

HISTPLOT
  </h2>
</div>

A histplot (histogram) is a plot that shows the distribution of frequency of values of one numerical variable. The X-axis of a histogram represents intervals (or “bins”) into which the range of values for the variable is divided. The Y-axis shows the frequency with which values fall into each interval. Histograms are useful for understanding how data is distributed, where the values are concentrated, whether there is symmetry, and if there are any outliers.

<b>NUMERICAL DATA</b>

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 8))
sns.set_style("darkgrid")

sns.histplot(data=tips, x='total_bill', hue='sex', kde=True, ax=axes[0],
             palette='coolwarm', element='step', fill=True, stat='density',
             common_norm=False, linewidth=1.5, edgecolor='black')
axes[0].set_title('Distribution of Total Bill by Gender', fontsize=14)
axes[0].set_xlabel('Total Bill', fontsize=12)
axes[0].set_ylabel('Density', fontsize=12)
axes[0].tick_params(axis='both', labelsize=10)

sns.histplot(data=tips, x='size', hue='time', kde=False, ax=axes[1],
             palette='pastel', element='bars', stat='count',
             linewidth=1.5, edgecolor='black', discrete=True)
axes[1].set_title('Distribution of Order Size by Time of Day', fontsize=14)
axes[1].set_xlabel('Order Size', fontsize=12)
axes[1].set_ylabel('Count', fontsize=12)
axes[1].tick_params(axis='both', labelsize=10)

plt.suptitle('Analysis of Distributions in the tips Dataset', fontsize=16, y=1.02)
plt.tight_layout(rect=[0, 0, 1, 0.98])

plt.show()
```

PIEPLOT

<b>Pieplot</b> is a good way to visualize the share of each category in the total amount of data. Suitable for a small number of categories.

<b>CATEGORICAL VARIABLE (IN A SMALL NUMBER OF FRACTIONS TO VISUALIZE THEM)</b>

```python
#make the data

year_counts = flights.groupby('year')['passengers'].sum()
```

```python
colors = sns.color_palette('viridis', len(year_counts))
plt.style.use('seaborn-v0_8-whitegrid')

plt.figure(figsize=(7, 7))
plt.pie(year_counts.values, labels = year_counts.index, autopct='%1.1f%%', colors=colors, 
        startangle=140, wedgeprops = { 'linewidth' : 2, 'edgecolor' : 'white' }, 
         textprops={'color': "black", 'fontweight': 'bold'}, shadow=True)
plt.title('Distribution passengers by years', color="#333", fontsize=20, fontweight='bold')
plt.show()
```

LINEPLOT
  </h2>
</div>

A lineplot is a type of plot that displays the change of one or more numerical variables with respect to another numerical variable (often time or some sequence). A lineplot connects data points with straight lines, making it easy to see trends, fluctuations, and the dynamics of changes in the data. This graph is especially useful for visualizing time series or sequential observations.

<b>TWO NUMERICAL DATA</b>

```python
sns.lineplot(x='day', y='tip', hue='sex', data=tips, errorbar='ci',
             marker='o', markersize=8, linestyle='-', linewidth=2,
             palette='coolwarm')
mean_tip = tips['tip'].mean()
plt.axhline(mean_tip, color='gray', linestyle='--', label=f'Mean Tip: ${mean_tip:.2f}')

plt.title('Average Tip Amount by Day and Gender', fontsize=14)
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Average Tip Amount', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, alpha=0.5)
plt.tight_layout()

plt.show()
```

COUNTPLOT

</div>

A countplot is a type of plot that shows the number of observations (frequency) for each category in a single column (feature). It displays the data as vertical bars, with the height of each bar corresponding to the number of times each value appears in the categorical column.

<b>CATEGORICAL VARIABLE</b>

```python
sex_counts = tips['sex'].value_counts(normalize=True) * 100

plt.figure(figsize=(6, 4))
sns.countplot(y='sex', data=tips, palette='deep', orient='h', edgecolor='black')
plt.xlabel('Count', fontsize=12)
plt.ylabel('Gender', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.title('Gender Distribution')

for i, p in enumerate(plt.gca().patches):
    width = p.get_width()
    height = p.get_height()
    x = p.get_x() + width - 10
    y = p.get_y() + height / 2
    plt.gca().annotate(f'{sex_counts.iloc[i]:.1f}%', (x, y),
                       ha='right', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

DISTPLOT
  </h2>
</div>

Distplot (distribution plot) is a function from the Seaborn library that combines several graphical elements to visualize the distribution of one numerical variable. It allows you to display a histogram (frequency distribution), a kernel density estimation (a smooth distribution estimate), and even a rug plot (individual “ticks” at the bottom of the plot, showing the location of individual values). Distplot provides a more comprehensive view of how the data is distributed than a simple histogram.

<b>ONE NUMERICAL VARIABLE</b>

```python
plt.figure(figsize=(8, 6))

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Гистограмма
sns.histplot(tips['total_bill'], ax=axes[0], color='#3498db', bins=20, edgecolor='white', alpha=0.7, kde=True)
axes[0].set_title('Histogram of Total Bill')
axes[0].set_xlabel('Total Bill')
axes[0].set_ylabel('Frequency')

# Кривая плотности
sns.kdeplot(tips['total_bill'], ax=axes[1], color='coral', alpha=0.6, linewidth=2)
axes[1].set_title('KDE of Total Bill')
axes[1].set_xlabel('Total Bill')
axes[1].set_ylabel('Density')

plt.tight_layout()
plt.show()
```
